### Reduction Operation

```python
e = torch.randint(size=(2,3),low=0,high=10,dtype=torch.float32)
e
```

```python
#sum
torch.sum(e)
torch.sum(e,dim=0)
torch.sum(e,dim=1)
```

```python
#mean
torch.mean(e)
torch.mean(e,dim=0)
torch.mean(e,dim=1)
```

```python
#median
torch.median(e)
torch.median(e,dim=0)
torch.median(e,dim=1)
```

```python
#max and min
torch.max(e)
# torch.max(e,dim=0)
# torch.max(e,dim=1)
torch.min(e)
```

```python
#product
torch.prod(e)
torch.prod(e,dim=0)
torch.prod(e,dim=1)
```

```python
#standard deviation
torch.std(e)
```

```python
#variance
torch.var(e)
```

```python
#argmax
torch.argmax(e)
torch.argmax(e,dim=0)
torch.argmax(e,dim=1)

#argmin
torch.argmin(e)
# torch.argmin(e,dim=0)
# torch.argmin(e,dim=1)
```

# Matrix Multiplication

```python
f = torch.randint(size=(3,2),low = 0 , high = 10)
g = torch.randint(size=(2,3),low = 0 , high = 10)
print(f)
print(g)
```

```python
#matrix multiplication
torch.matmul(f,g)
```

```python
vector1 = torch.tensor([1,2])
vector2 = torch.tensor([3,4])

#dot product
torch.dot(vector1,vector2)
```

```python
#transpose
torch.transpose(f,0,1)
```

```python
h = torch.randint(size=(3,3),low = 0 , high = 10 , dtype=torch.float64)
h
```

```python
#determinant
torch.det(h)
```

```python
#inverse
torch.inverse(h)
```

## Comparision Operation

```python
i = torch.randint(size=(3,3),low = 0 , high = 10)
j = torch.randint(size=(3,3),low = 0 , high = 10)
print(i)
print(j)
```

```python
#greater than
i>j

#less than
i<j

#equal to
i == j

#not equal to
i != j

#greater than equal to
i>=j

#less than equal to
i<=j
```

# Special Function
