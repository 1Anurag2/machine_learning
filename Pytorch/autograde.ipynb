# Autograd

Autograd is an automatic differentiation system that computes gradients of tensors for backpropagation in deep learning models.

```python
import torch
```

```python
x = torch.tensor(3.0,requires_grad = True)
```

```python
y = x**2       # to calculate dy/dx of y =  x^2
```

```python
x
```

```python
y
```

```python
y.backward()
```

```python
x.grad
```

# Example

```python
import math
def dz_dx(x):
  return 2*x * math.cos(x**2)
```

```python
dz_dx(3)
```

```python
#using torch to calculate differntiation of dz/dz   where  y = x^2  and z = siny

x = torch.tensor(3.0,requires_grad = True)
y = x**2
z = torch.sin(y)
z.backward()
```

```python
x.grad
```

# Simple Nural Network

```python
x = torch.tensor(6.7)
y = torch.tensor(0.0)
```

```python
w = torch.tensor(1.0,requires_grad = True)
b = torch.tensor(0.0,requires_grad = True)
```

```python
w
```

```python
b
```

```python
z = w*x + b
z
```

```python
y_pred = torch.sigmoid(z)
y_pred
```

```python
import torch.nn.functional as F
loss = F.binary_cross_entropy(y_pred,y)
loss
```

```python
loss.backward()
```

```python
print(w.grad)
print(b.grad)
```

# Clearing Grad

```python
x = torch.tensor(2.0,requires_grad = True)
x
```

```python
y = x**2
y.backward()
x.grad
```

```python
x.grad.zero_()
```
