```python
import matplotlib.pyplot as plt
x = ["Mon","Tue","Wed","Thru","Fri","Sat","Sun"]
y = [10,25,80,34,22,89,12]

plt.plot(x,y,color="blue",linestyle="--",linewidth=2, marker='o',label="2025 sales Data")
plt.title("Sales Information of Week")
plt.xlabel("Days name")
plt.ylabel('Sales Number')
plt.legend(loc='upper left',fontsize=12)
plt.grid(color = 'gray',linestyle=':',linewidth=1)
plt.xlim(1,4)
plt.ylim(0,70)
plt.xticks([1,2,3,4,5,6,7],['D1','D2','D3','D4','D5','D6','D7'])
plt.show()
```

```python
import matplotlib.pyplot as plt
Month = ["Jan","Feb",'Mar','Apr','May','June','July','Aug','Sep','Oct','Nov','Dec']
profit = [1000,45000,7800,46388,67885,46483,5347,78456,875,7483,8946,4536]
plt.bar(Month,profit,color="orange",label="Profit Data")
plt.title("Full Year Profit Data")
plt.xlabel('Months')
plt.ylabel('Profits')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
#plt.pie(values,labels=label_list,colors=color_list,autopct='%1.1f%%')
regions = ['North','South','East','West']
revenue = [3000,2000,1500,1000]
plt.pie(revenue,labels=regions,autopct='%1.1f%%',colors=['gold','skyblue','lightgreen','coral'])
plt.title('Revenue Contribution By Region')
plt.show()
```

```python
#plt.hist(data,bins=number_of_bins,color='colorname',edgecolor='black')
import matplotlib.pyplot as plt
scores = [23,56,78,90,23,32,33,45,67,89,24,67,89,14,90]
plt.hist(scores,bins=7,color="purple",edgecolor='black')
plt.xlabel('Score Range')
plt.ylabel("Number of Students")
plt.show()
```

```python
#plt.scatter(x,y,color='color_name',marker='marker style',label='label name')
import matplotlib.pyplot as plt
hours_studies = [1,2,3,4,5,6,7,8]
exam_scores = [50,55,60,70,75,80,90,100]
plt.scatter(hours_studies,exam_scores,color='green',marker='o',label="Student Data")
plt.xlabel('Hours studied')
plt.ylabel('Exam Score')
plt.title('Relationship between time and score')
plt.legend()
plt.grid(True)
plt.show()
```
